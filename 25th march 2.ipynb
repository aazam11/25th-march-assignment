```python
# Load the Google Playstore dataset
file_path = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'
playstore_data = pd.read_csv(file_path)

# Task 1: Examine dimensions of the dataset
rows, columns = playstore_data.shape
print(f"The dataset has {rows} rows and {columns} columns.")
```

```python
# Task 2: Rating variation by category - Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Rating', data=playstore_data)
plt.title('Rating Variation by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
# Task 3: Check for missing values
missing_values = playstore_data.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
# Task 4: Relationship between size and rating - Scatter plot
plt.figure(figsize=(10,50))
sns.scatterplot(x='Rating', y='Size', data=playstore_data)
plt.title('Relationship between App Size and Rating')
plt.xlabel('Rating')
plt.ylabel('Size')
plt.show()
```

```python
#task 5
import matplotlib.pyplot as plt
# Load the Google Playstore dataset
file_path = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'
playstore_data = pd.read_csv(file_path)

# Clean the 'Price' column and convert it to numeric
playstore_data['Price'] = pd.to_numeric(playstore_data['Price'].str.replace('$', '').replace('Everyone', '10'))

# Bar chart for average prices by app type
plt.figure(figsize=(10, 6))
average_prices = playstore_data.groupby('Type')['Price'].mean()
sns.barplot(x=average_prices.index, y=average_prices.values)
plt.title('Average Prices by App Type')
plt.xlabel('Type')
plt.ylabel('Average Price')
plt.show()
```
